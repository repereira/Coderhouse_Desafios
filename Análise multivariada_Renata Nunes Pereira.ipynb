### Histograma

### Avaliando a largura da pétala:

```python
# Agrupar por categoria e contar a frequência dos valores
df_grouped = df.groupby('species')['petal_width'].value_counts().unstack().fillna(0)

# Plotar o histograma
df_grouped.plot(kind='bar', stacked=False, figsize=(10, 10))
plt.xlabel('species')
plt.ylabel('Frequencia')
plt.title('Histograma por Species')
plt.legend(title='petal_width')
plt.show()
```

### Comprimento da pétala

```python
# Agrupar por categoria e contar a frequência dos valores
df_grouped = df.groupby('species')['petal_length'].value_counts().unstack().fillna(0)

# Plotar o histograma
df_grouped.plot(kind='bar', stacked=False, figsize=(20, 15))
plt.xlabel('species')
plt.ylabel('Frequencia')
plt.title('Histograma por Species')
#plt.legend(title='petal_length')
plt.show()
```
