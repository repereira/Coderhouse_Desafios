### 1. Carregar o arquivo. Executar estatística descritiva básica no Python.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('10 - HRDataset_v14.csv', encoding='ISO-8859-1')
df
```

```python
df.describe()
```

### 2. Fazer um histograma com salários. Quais faixas salariais são as mais populares?

```python
sns.histplot(df['Salary'], kde = True)
```

```python
salario_medio = df['Salary'].mean()
salario_medio = np.round(salario_medio ,decimals=2)
salario_medio 
```

```python
salario_desvio = df['Salary'].std()
salario_desvio = np.round(salario_desvio ,decimals=2)
salario_desvio
```

```python
faixa_salarial_max= salario_medio + salario_desvio
faixa_salarial_min = salario_medio - salario_desvio
print(f'A faixa salarial que mais se repete varia entre {faixa_salarial_min:.2f} e {faixa_salarial_max:.2f}')
```

### 3. Fazer um gráfico de violino com salários, discriminados por gênero.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='GenderID', y='Salary', data=df, palette='Set2')
plt.title('Distribuição de Salário por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Salário')
plt.show()
```

### 4. Traçar a série temporal correspondente à data de contratação (DateofHire).

```python
df.info()
```

```python
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
```

```python
df.info()
```

```python
df['Ano'] = df['DateofHire'].dt.year
df
```

```python
df_por_ano = df.groupby('Ano').mean()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_por_ano.index, df_por_ano['Salary'], marker='o', linestyle='-', label='Média de Salários')
plt.title('Série Temporal de ano de Contratação')
plt.xlabel('Ano de Contratação')
plt.ylabel('Salário médio')
plt.grid(True)
plt.tight_layout()
plt.show()
```
