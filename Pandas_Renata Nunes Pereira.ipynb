1. Construir um objeto de Série cronológica com o índice igual ao ano dos jogos olimpicos.

```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding='ISO-8859-1')
data
```

```python
medalhas_por_ano = data.groupby('Year').size()
```

```python
# Criar a série cronológica com o índice igual ao ano
serie_cronologica = pd.Series(medalhas_por_ano, index=medalhas_por_ano.index)

serie_cronologica
```

4. Construir uma tablea mostrando quantas medalhas foram ganhas pelos homens no total em cada ano em que evento foi realizado

```python
# Filtrar os dados para contar apenas as medalhas ganhas pelos homens
medalhas_homens = data[data['Gender'] == 'Men']

# Contar o número de medalhas ganhas pelos homens em cada ano
medalhas_homens_por_ano = medalhas_homens.groupby('Year').size().reset_index(name='Total Medalhas')

# Exibir o resultado
medalhas_homens_por_ano
```
